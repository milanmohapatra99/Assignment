```python
#Importing Required Libraries
import pandas as pd
import numpy as np
```

```python
#Reading Data From CSV file in data1
data1 = pd.read_csv('recordsmatchingtask - data1.csv')
```

```python
#Creating Unique column ID - "Data1_UniqueID"
data1['Data1_UniqueID'] = data1['Order ID']+data1['Product ID']
```

```python
# Viewing Record of Data1 
data1
```

```python
#Reading Data From Data2 csv file
data2 = pd.read_csv('recordsmatchingtask - data2.csv')
```

```python
#Creating Unique column ID - "Data2_UniqueID"
data2['Data2_UniqueID'] = data2['Order ID']+data2['Product ID']
```

```python
# Viewing Data2
data2
```

```python
# Two data sets are merged on the common Unique ID
final_data = data1.merge(data2, how='outer', left_on=['Data1_UniqueID'], right_on=['Data2_UniqueID'])
final_data
```
